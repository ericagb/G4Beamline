```python
num = 100
beam = [[None for i in range(num)] for j in range(1)]
mu = [[None for i in range(num)] for j in range(1)]
nEv = [100000]
angle = [f"{((x+1)*0.0005):.4f}" for x in range(0,100)]
for i in range(1):
    for j in range(num):
        beam[i][j] = plot.add_text_file(f"nEv100000|angle{angle[j]}.txt")
        mu[i][j] = plot.get_particle_count(beam[i][j],"mu-")


mu = np.asarray(mu)
angle = np.array(angle,dtype=float)
nEv = np.asarray(nEv, dtype=float)

divider = nEv / 100000
for i in range(1):
    for j in range(num): 
        mu[i][j] = (angle[j]/0.001)**2 * mu[i][j] / divider[i]

angle = [f"{((x+1)*0.0005):.4f}" for x in range(0,100)]

df = pd.DataFrame(np.c_[mu[0]], index=angle)
ax = df.plot(kind='bar', rot=45, width=0.8)

for container in ax.containers:
    plt.bar_label(container, label_type='edge',rotation=90, padding=5)
plt.legend(['1e5 nEv', '1e6 nEv' , '1e7 nEv'],loc='upper left')

plt.savefig("histo_normal_count.pdf", dpi=800)
plt.show()
#plot.save_figure(ax,"histo_normal.pdf")
```

```python
angle = [float(item)*1000 for item in angle]
plt.scatter(angle, mu[0,:])
plt.xlabel("Initial Pion Distribution (mrad)")
plt.ylabel("Count of Muons")
plt.title("Muon Count vs Initial Pion Distribution")
plt.show()
```
